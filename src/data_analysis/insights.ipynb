```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def plot_feature(file, name, deg, title, title2, plot_title): 
    res = pd.read_csv(file)
    res = res[name].tolist()

    res_x = np.arange(0,len(res))
    res_poly = np.poly1d(np.polyfit(res_x, res, deg))

    plt.plot(res_x, res_poly(res_x),c="#ff00ff", label=title)
    plt.plot(res_x, res,c="#0000ff", label=title2)
    
    plt.title(title)
    plt.legend()
    plt.show()

def plot_double(file, name, deg, title, title2, file2, name2, deg2, ti, ti2, plot_title): 
    res = pd.read_csv(file)
    res = res[name].tolist()

    res_x = np.arange(0,len(res))
    #res_poly = np.poly1d(np.polyfit(res_x, res, deg))

    #plt.plot(res_x, res_poly(res_x),c="#ff00ff", label=title)
    plt.plot(res_x, res,c="#0000ff", label=title2)

    res2 = pd.read_csv(file2)
    res2 = res2[name2].tolist()

    res2_x = np.arange(0,len(res2))
    #res2_poly = np.poly1d(np.polyfit(res2_x, res2, deg2))

   # plt.plot(res2_x, res2_poly(res2_x),c="#ff00ff", label=ti)
    plt.plot(res2_x, res2,c="#ff00ff", label=ti2)
    
    
    plt.title(title)
    plt.legend()
    plt.show()
```
